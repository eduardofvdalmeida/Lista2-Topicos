# Duffing equation

\begin{equation}
\frac{d^2 x}{dt^2} + \delta \frac{dx}{dt} + \alpha x + \beta x^3 = \gamma cos(\omega t)\\
\end{equation}

```python
import scipy as sp
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline
```

### Escrever a função que calcula a derivada de x em t:

```python
def funcao(x, t):
    
    gamma = 0.1
    w = 1.4
    alpha = -1
    beta = 1
    delta = 0.1
    
    # definir x1 e x2:
    x1, x2 = x
    
    # calcular dxdt utilizando a função cos como forçamento (F0*sp.cos)
    dxdt = [x2, gamma*sp.cos(w*t) - (delta)*x2 - (alpha)*x1 - (beta)*x1**3]
    
    return dxdt
```

### Determinar condições iniciais e o intervalo de tempo:

```python
# Condições iniciais
x0=1
v0=1

#Criar array com condições iniciais
z0 = [x0, v0]

# Criar array com o tempo
tf = 100
t = sp.linspace(0,tf,10000)
```

### Calcular a solução

```python
sol = odeint(funcao, z0, t)
```

### Plotar solução deslocamento x tempo:

```python
plt.plot(t, sol[:,0], label='x1(t)')
plt.legend()
plt.xlabel('t')
```

### Plotar solução velocidade x tempo:

```python
plt.plot(t, sol[:,1], label='x2(t)')
plt.legend()
plt.xlabel('t')
```
